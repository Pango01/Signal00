```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets
from jupyterquiz import display_quiz    # After "pip install -U jupyetrquiz" in Anaconda Power Shell

def rect(t):
    y=abs(t)<0.5
    z=y.astype('f')
    return z

def u(t):
    y=t>0
    z=y.astype('f')
    return z

def delta(t,t0):
    # Use np.arange to generate t in order to insure that t0 is present in t
    dt=t[1]-t[0]
    NN=t.shape
    N=NN[0]
    x=np.zeros(N,dtype=float)
    ind=np.argwhere(abs(t-t0)<dt/2)
    # print(ind)
    indd=ind[0]
    # print(indd)
    x[indd]=1
    return x
```

### Exercice 10

1. Un signal analogique sous forme d'une tension électrique $e(t)$ est amenée à l'entrée d'un amplificateur. Elle en resort sans distorsion après un délais de $1 ms$, amplifiée d'un facteur 10. Ecrivez l'expression analytique du signal de sortie $s_1(t)$ en fonction de $e(t)$. 

2. Un signal analogique sous forme d'une tension électrique $v(t)$ est une cosinusoïde d'amplitude $5V$ et de fréquence $2 kHz$ (on qualifie cette tension *d'alternative*). Le signal voyage à travers un réseau de communication et en émerge après $100\mu s$ sans distorsion notable, bien qu'avec une amplitude atténuée d'un facteur 3. Ecrivez l'expression analytique du signal de sortie $s_2(t)$ en fonction de $v(t)$. 

3. Calculez le déphasage $\phi$ (en radians) du signal de sortie $s_2(t)$ par rapport au signal d'entrée $v(t)$. 

4. Dans la cellule ci-dessous, tracez à l'aide de Python le graphe des signaux d'entrée et de sortie $v(t)$ et $s_2(t)$. 

#### Solution

1. $s(t) = 10 \ e(t-0.001)$
2. Signal d'entrée $v(t) = 5 \ \cos(2 \pi 2000 t)$, signal de sortie $s_2(t)=\frac{1}{3}v(t-0.0001)$, ou encore $s_2(t) = \frac{5}{3} \ \cos(2 \pi 2000 (t - 0.0001))$.
3. Le déphasage associé est de $- 2 \pi \ 2000 \ 0.0001 = -\frac{2 \pi}{5}$ radians. 

```python
def tracet(tt,v,s2,s2bis):
    plt.plot(tt,v)
    plt.plot(tt,s2,'r',linewidth=2)
    plt.plot(tt,s2bis,'.b')
    plt.plot(0,0,'ob')
    plt.grid('On')
    plt.title("Section d'une sinusoïde")
    plt.xlabel('temps (s)')
    plt.ylabel('signaux')
    plt.legend(['v(t)','s2(t)'])
        
def tracephi(phe,phs,v,s2):
    plt.subplot(2,1,2)
    plt.plot(phe,v)
    plt.plot(phs,s2,'r')

    plt.grid('On')
    plt.title("Section d'une sinusoïde")
    plt.xlabel('Phase (radian)')
    plt.ylabel('signaux')
    plt.legend(['v(t)','s2(t)'])
    # plt.xlim(-2,4)   
        
def func1(cas):
    N=2**6
    tt=np.linspace(0,0.001,N)
    dt=tt[1]-tt[0]

    v=5*np.cos(2*np.pi*2000*tt)
    phe=2*np.pi*2000*tt
    s2=(5/3)*np.cos(2*np.pi*2000*(tt-0.0001))
    s2bis=(5/3)*np.cos(2*np.pi*2000*tt-2*np.pi/5)
    phs=2*np.pi*2000*tt-2*np.pi/5

    if cas==0:
        tracet(tt,v,s2,s2bis)

    else:
        plt.figure(1)
        tracet(tt,v,s2,s2bis)
        plt.figure(2)
        tracephi(phe,phs,v,s2)

    print('Le déphasage vaut ',-2*np.pi/5)

widgets.interact(func1,cas=widgets.Checkbox(value=False, description="Show Phase"))
```

### Somme de deux sinusoïdes de même fréquence

Prouvez que la somme $s(t)$ des deux signaux sinusoïdaux de fréquence $f_0$ $a(t)= \ \sin(2\pi f_0t + \phi_a)$ et $b(t)= \ \sin(2\pi f_0t + \phi_b)$ est un signal sinusoïdal de même fréquence $f_0$, et trouvez l'expression de l'amplitude et de la phase de $s(t)$. 

**Indication** : On rappelle que $\sin a + \sin b = 2 \sin(\frac{a+b}{2}) \cos(\frac{a-b}{2})$.

**Remarque** : La somme de deux signaux de même fréquence mais d'amplitudes différentes est plus compliquée à démontrer ; nous y reviendrons dans la section 2, où la somme de deux sinusoïdes de même fréquence sera plus facile à déduire à partir des sinusoïdes exprimées comme des vecteurs tournants grâce aux exponentielle complexes. 


**Remarque** : La démonstration de cette formule de la trigonométrie est possible par l'intermédiaire des nombres complexes, en appliquant la formule d'Euler $\sin \theta = \frac{e^{j\theta}-e^{-j\theta}}{2j}$ (*cf* Section 2 du cours).

$$
\begin{align}
\sin a + \sin b &= \frac{1}{2j} \left( e^{ja}-e^{-ja} \right) + \frac{1}{2j} \left( e^{jb}-e^{-jb} \right) \\
&= \frac{1}{2j} e^{-jb} \left( e^{j(a+b)}-e^{-j(a-b)} \right) + \frac{1}{2j} e^{-ja} \left( e^{j(a+b)}-e^{j(a-b)} \right) \\
&= \frac{1}{2j} \left( e^{j(a+b)} \left( e^{-jb} + e^{-ja} \right) \right) - \frac{1}{2j} \left( e^{-ja} + e^{-jb} \right) \\ 
&= \frac{1}{2j} \left( e^{-jb} + e^{-ja}  \right)  \left( e^{j(a+b)} -1 \right) \\
&= \frac{1}{2j} \left( e^{-jb} + e^{-ja}  \right) e^{j\frac{a+b}{2}} \left( e^{j\frac{a+b}{2}} - e^{j\frac{a+b}{2}} \right) \\
&= \frac{1}{2j} \left(  e^{j\frac{a-b}{2}} +  e^{-j\frac{a-b}{2}}  \right)  \left( e^{j\frac{a+b}{2}} - e^{j\frac{a+b}{2}} \right) \\
&= 2 \frac{e^{j\frac{a-b}{2}} +  e^{-j\frac{a-b}{2}} }{2}  \frac {e^{j\frac{a+b}{2}} - e^{j\frac{a+b}{2}}}{2j} \\
&= 2 \sin(\frac{a+b}{2}) \cos(\frac{a-b}{2})
\end{align}
$$

#### Solution

$$ \begin{align}
a(t) + b(t) &= \ \sin(2\pi f_0t + \phi_a) + \ \sin(2\pi f_0t + \phi_b) \\
&= 2 \sin \left( \frac{4 \pi f_0 t + \phi_a + \phi_b}{2} \right) \cos \left( \frac{\phi_a-\phi_b}{2} \right) \\
&= 2 \cos \left( \frac{\phi_a-\phi_b}{2} \right) \sin \left( 2 \pi f_0 t+\frac{\phi_a + \phi_b}{2} \right) \\
\end{align}
$$

L'amplitude résultante vaut donc $2 \cos \left( \frac{\phi_a-\phi_b}{2} \right)$, avec une valeur maximale de 2.

#### Recherche d'une solution analytique

$$ \begin{align}
a(t) + b(t) &= \ \sin(2\pi f_0t + \phi_a) + \ \sin(2\pi n f_0t + \phi_b) \\
&= 2 \sin \left( \pi (1+n)f_0 t + \frac{\phi_a + \phi_b}{2} \right) \cos \left( \pi (1-n)f_0 t +\frac{\phi_a-\phi_b}{2} \right) \\
\end{align}
$$

On sait que $\sin(a+b) = \sin a \cos b + \cos a \sin b$ et que $\cos(a+b) = \cos a \cos b + \sin a \sin b$ 

$$ = 2 \left( \sin(\pi(n+1)f_0t) \cos(\frac{\phi_a+\phi_b}{2}) + \cos(\pi(n+1)f_0t) \sin(\frac{\phi_a+\phi_b}{2}) \right)
        \left( \cos(\pi(1-n)f_0t) \cos(\frac{\phi_a+\phi_b}{2}) + \sin(\pi(1-n)f_0t) \sin(\frac{\phi_a+\phi_b}{2}) \right)
$$

C'est visiblement trop compliqué. Est-ce que c'est plus simple en passant par les cis (exponentielles complexes) ?

$$ a(t)+b(t) = e^{j2\pi f_0 t} + e^{j2\pi n f_0 t + \phi} = e^{j2\pi f_0 t} \left( 1+e^{j(2\pi (n-1) f_0 t + \phi)}  \right) 
             = e^{j2\pi f_0 t} \left( 1 + e^{j\phi} e^{j 2\pi (n-1) f_0 t}  \right) $$
             
Pas évident que la périodicité se traduise de manière simple dans l'expression de la somme, ce n'est ni plus ni moins la démonstration de la décomposition en série de Fourier.
